```python
import pandas as pd

df = pd.read_csv("../data/state_month_full.csv")
df.head(5)
```

```python
# drop state_name
df = df.drop(columns=["state_name"], errors="ignore")

# NaN count per column (sorted)
nan_counts = df.isna().sum().sort_values(ascending=False)
print(nan_counts)

# (optional) also show % missing
nan_pct = (df.isna().mean() * 100).sort_values(ascending=False)
print("\n% missing:")
print(nan_pct)
```

```python
cols = ["MSPUS", "PPIFDG", "CPI_YOY"]

# first 10 values (keeps NaNs)
print(df[cols].head(50).to_string(index=False))
```

```python
# drop MSPUS and PPIFDG
df = df.drop(columns=["MSPUS", "PPIFDG"], errors="ignore")
```

```python
df_clean = df.dropna(subset=["CPI_YOY"]).copy()
```

```python
df_clean.shape
```

```python
# NaN count per column (sorted)
nan_counts = df_clean.isna().sum().sort_values(ascending=False)
print(nan_counts)
```

```python
import os

os.makedirs("../data", exist_ok=True)
out_path = os.path.join("../data", "cleaned_dataset.csv")

df_clean.to_csv(out_path, index=False)

print("Saved:", out_path, "Rows:", len(df_clean), "Cols:", df_clean.shape[1])
```
